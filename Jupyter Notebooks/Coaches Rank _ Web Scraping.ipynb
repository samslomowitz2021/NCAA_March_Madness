```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd

import requests
```

```python
url = "https://www.espn.com/mens-college-basketball/story/_/id/33253706/ranking-2022-ncaa-tournament-men-basketball-coaches-players-1-68"
```

```python
response = requests.get(url)

html = response.text
html[0:500]

soup = BeautifulSoup(html, "lxml")
```

```python


# print(soup.prettify())
```

```python
print(soup.title.text)
```

```python
value = soup.find('strong').text
print(value)
```

```python
coaches=[]
for coach in soup.find_all("strong"):
    coach= coach.text.split(',')
  
    
    coaches.append(coach)
```

```python
coaches
```

```python
del coaches[0:7]
coaches
```

```python
df = pd.DataFrame(coaches, columns =['Rank', 'team']) 
df
```

```python
df2=df['Rank'].str.split(' ', expand=True).rename(columns={0:"Rank", 1:"FirstName", 2:"LastName"})
df2
```

```python
df3=df2['Rank'].str.split('.', expand=True).rename(columns={0:"Rank"})
df3 = df3.drop(df3.columns[[1]], axis=1)
df3
```

```python
df = pd.concat([df3['Rank'], df2['FirstName'],df2['LastName'], df['team']], axis=1)
df
```

```python
df.to_csv('Coaches.csv')
```
